```python
# d_dimorphos = 27139*1000 # m, v_inf*t_dimorphos, pixels at which ast is first visible in camera per nasa final report
d_dimorphos = 18430*1000 # m, pixels at which SmartNav can target Dimorphos directly per final report
# print(d_dimorphos)
dia_dimorphos = 151 # m
# angular diameter formula from Wikipedia lol https://en.wikipedia.org/wiki/Angular_diameter#For_a_sphere
ang_dia_dimorphos = 2*asin(dia_dimorphos/(2*d_dimorphos)) #radians, 
ifov = 2.5e-6 #murad
print("Number of pixels visible by Dimorphos",ang_dia_dimorphos/ifov)

dia_pdc25 = 150
v_inf = 8.45
d_pdc25 = ( dia_pdc25/(2*sin(ang_dia_dimorphos/2)) )
t_pdc25 = d_pdc25/(1000*60*v_inf)
print("Time to when SmartNav switch to target PDC25 directly",t_pdc25)

draco_fov = 0.29*pi/180 # fov in radians
print(draco_fov/ifov)

# Need
pp_cis2521f = 6.5 # pixel pitch, microns
pp_emeraldgen2 = 2.8 # pixel pitch, microns
ifov_emeraldgen2 = (pp_emeraldgen2/pp_cis2521f) * ifov
print("New IFOV of Teledyne Emerald sensor",ifov_emeraldgen2)

#pixels = ang_dia_dimorphos/ifov = ang_dia_pdc25_emerald/ifov_emeraldgen2

ang_dia_pdc25_emerald = (ang_dia_dimorphos/ifov) * ifov_emeraldgen2
d_pdc25_emerald = ( dia_pdc25/(2*sin(ang_dia_pdc25_emerald/2)) )
t_pdc25_emerald = d_pdc25_emerald/(1000*60*v_inf)
print("Time to when SmartNav switch to target PDC25 directly, Emerald sensor",t_pdc25_emerald)
```
